```python
features_df = spark.read.format("csv").option("header","false").schema(schema).option("delimiter","\t").load(output_loc)
features_pd = features_df.toPandas()
    
percentile_columns_label = ['(frame_len)', '(ip_len)', '(ip_ttl)', '(tcp_len)', '(tcp_winsize)']

percentile_columns = ['percentiles(1_2_5_10_15_20_25_50_75_90_95_97_99_frame_len)'\
                     , 'percentiles(1_2_5_10_15_20_25_50_75_90_95_97_99_(ip_len)'\
                     , 'percentiles(1_2_5_10_15_20_25_50_75_90_95_97_99_(ip_ttl)'\
                     , 'percentiles(1_2_5_10_15_20_25_50_75_90_95_97_99_(tcp_len)'\
                     , 'percentiles(1_2_5_10_15_20_25_50_75_90_95_97_99_(tcp_winsize)']

tcp_app_port_label = ['(tcp_dstport)', '(tcp_dstport)', '(tcp_srcport)', '(tcp_srcport)']
tcp_app_ports = ['num_appport_ranges(tcp_dstport)'\
                , 'frac_appport_ranges(tcp_dstport)'\
                , 'num_appport_ranges(tcp_srcport)'\
                , 'frac_appport_ranges(tcp_srcport)']

udp_app_port_label = ['(udp_dstport)', '(udp_dstport)', '(udp_srcport)', '(udp_srcport)']
udp_app_ports = ['num_appport_ranges(udp_dstport)'\
                , 'frac_appport_ranges(udp_dstport)'\
                , 'num_appport_ranges(udp_srcport)'\
                , 'frac_appport_ranges(udp_srcport)']

features_pd['num_icmp_tcp_udp_gre(ip_proto)'] = features_pd['num_icmp_tcp_udp_gre(ip_proto)'].str.strip('[]')
features_pd['num_icmp(ip_proto)'], features_pd['num_tcp(ip_proto)'], features_pd['num_udp(ip_proto)'], features_pd['num_gre(ip_proto)'] = features_pd['num_icmp_tcp_udp_gre(ip_proto)'].str.split(',').str
features_pd.drop(columns=['num_icmp_tcp_udp_gre(ip_proto)'], inplace=True)

features_pd['frac_icmp_tcp_udp_gre(ip_proto)'] = features_pd['frac_icmp_tcp_udp_gre(ip_proto)'].str.strip('[]')
features_pd['frac_icmp(ip_proto)'], features_pd['frac_tcp(ip_proto)'], features_pd['frac_udp(ip_proto)'], features_pd['frac_gre(ip_proto)'] = features_pd['frac_icmp_tcp_udp_gre(ip_proto)'].str.split(',').str
features_pd.drop(columns=['frac_icmp_tcp_udp_gre(ip_proto)'], inplace=True)

features_pd['num_echo_reply_destination_unreachable_redirect_message_echo_request(icmp_type)'] = features_pd['num_echo_reply_destination_unreachable_redirect_message_echo_request(icmp_type)'].str.strip('[]')
features_pd['num_echo_reply(icmp_type)'], features_pd['num_destination_unreachable(icmp_type)'], features_pd['num_redirect_message(icmp_type)'], features_pd['num_echo_request(icmp_type)'] = features_pd['num_echo_reply_destination_unreachable_redirect_message_echo_request(icmp_type)'].str.split(',').str
features_pd.drop(columns=['num_echo_reply_destination_unreachable_redirect_message_echo_request(icmp_type)'], inplace=True)

features_pd['frac_echo_reply_destination_unreachable_redirect_message_echo_request(icmp_type)'] = features_pd['frac_echo_reply_destination_unreachable_redirect_message_echo_request(icmp_type)'].str.strip('[]')
features_pd['frac_echo_reply(icmp_type)'], features_pd['frac_destination_unreachable(icmp_type)'], features_pd['frac_redirect_message(icmp_type)'], features_pd['frac_echo_request(icmp_type)'] = features_pd['frac_echo_reply_destination_unreachable_redirect_message_echo_request(icmp_type)'].str.split(',').str
features_pd.drop(columns=['frac_echo_reply_destination_unreachable_redirect_message_echo_request(icmp_type)'], inplace=True)

for column,label in zip(percentile_columns, percentile_columns_label):
    features_pd[column] = features_pd[column].str.strip('[]')
    features_pd['p1'+label], features_pd['p2'+label], features_pd['p5'+label], features_pd['p10'+label], features_pd['p15'+label], features_pd['p20'+label], features_pd['p25'+label], features_pd['p50'+label], features_pd['p75'+label], features_pd['p90'+label], features_pd['p95'+label], features_pd['p97'+label], features_pd['p99'+label] = features_pd[column].str.split(',').str
    features_pd.drop(columns=[column], inplace=True)

for column,label in zip(tcp_app_ports, tcp_app_port_label):
    if 'num_' in column:
        prefix = 'num_'
    else :
        prefix = 'frac_'
    features_pd[column] = features_pd[column].str.strip('[]')
    features_pd[prefix+'well_known'+label], features_pd[prefix+'registered_ports'+label], features_pd[prefix+'dynamic_ports'+label], features_pd[prefix+'remote_access_tcp'+label], features_pd[prefix+'mail_tcp'+label], features_pd[prefix+'networking_tcp'+label], features_pd[prefix+'document_retrieval_tcp'+label] = features_pd[column].str.split(',').str
    features_pd.drop(columns=[column], inplace=True)

for column,label in zip(udp_app_ports, udp_app_port_label):
    if 'num_' in column:
        prefix = 'num_'
    else :
        prefix = 'frac_'
    features_pd[column] = features_pd[column].str.strip('[]')
    features_pd[prefix+'well_known'+label], features_pd[prefix+'registered_ports'+label], features_pd[prefix+'dynamic_ports'+label], features_pd[prefix+'mail_udp'+label], features_pd[prefix+'networking_udp'+label], features_pd[prefix+'document_retrieval_udp'+label] = features_pd[column].str.split(',').str
    features_pd.drop(columns=[column], inplace=True)
features_pd.to_csv('darknet_features_201901_06.csv', index=False)
```
